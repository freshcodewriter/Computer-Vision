```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import matplotlib as mpl
```

```python
dummy = np.arange(-2,2,0.1)
dummy_meshgrid = np.array(np.meshgrid(dummy,dummy))
w = dummy_meshgrid[0,:,:].reshape(-1,1)
b = dummy_meshgrid[1,:,:].reshape(-1,1)
x=1
def Sigmoid(w,b):
    tmp = -(w*x + b)
    loss = 1 / (1 + np.exp(tmp))
    return loss
Sigmoid_loss = Sigmoid(w,b)
fig = plt.figure()
ax = plt.axes(projection='3d')
zdata = Sigmoid_loss[:,0]
xdata = w[:,0]
ydata = b[:,0]
ax.scatter3D(xdata, ydata, zdata, c=zdata)
ax.set_xlabel('Weight')
ax.set_ylabel('Bias')
ax.set_zlabel('loss')
p = ax.scatter(xdata, ydata, zdata, c=zdata)
fig.colorbar(p,pad=0.07)
plt.title('Sigmoid Function')
plt.savefig('Sigmoid Function')
```

```python
def l2(Sigmoid_loss,true_y):
    l2_loss = 0.5*(Sigmoid_loss - true_y) * (Sigmoid_loss - true_y)
    return l2_loss
true_y = np.full((1600, 1), 0.5)
l2_loss = l2(Sigmoid_loss,true_y)
fig = plt.figure()
ax = plt.axes(projection='3d')
zdata = l2_loss[:,0]
xdata = w[:,0]
ydata = b[:,0]
ax.scatter3D(xdata, ydata, zdata, c=zdata)
ax.set_xlabel('Weight')
ax.set_ylabel('Bias')
ax.set_zlabel('loss')
p = ax.scatter(xdata, ydata, zdata, c=zdata)
fig.colorbar(p,pad=0.07)
plt.title('l2 loss')
plt.savefig('l2 loss')
```

```python
def backprop_l2(w,b,true_y):
    tmp_1 = np.exp(-(w*x + b))
    loss = (1 / (1 + tmp_1)) - true_y
    tmp_2 = -1 / ((tmp_1 + 1)**2)
    df_w = tmp_1*tmp_2*(-x)*loss
    return df_w
true_y = np.full((1600, 1), 0.5)
df_w = backprop_l2(w,b,true_y)
fig = plt.figure()
ax = plt.axes(projection='3d')
zdata = df_w[:,0]
xdata = w[:,0]
ydata = b[:,0]
ax.scatter3D(xdata, ydata, zdata, c=zdata)
ax.set_xlabel('Weight')
ax.set_ylabel('Bias')
ax.set_zlabel('loss')
p = ax.scatter(xdata, ydata, zdata, c=zdata)
fig.colorbar(p,pad=0.07)
plt.title('gradient of l2 loss')
plt.savefig('gradient of l2 loss')
```

```python
def cross_entropy(Sigmoid_loss,true_y):
    tmp_1 = true_y * np.log(Sigmoid_loss)
    tmp_2 = (1-true_y) * np.log(1 - Sigmoid_loss)
    cross_entropy_loss = -(tmp_1 + tmp_2)
    return cross_entropy_loss
true_y = np.full((1600, 1), 0.5)
cross_entropy_loss = cross_entropy(Sigmoid_loss,true_y)
fig = plt.figure()
ax = plt.axes(projection='3d')
zdata = cross_entropy_loss[:,0]
xdata = w[:,0]
ydata = b[:,0]
ax.scatter3D(xdata, ydata, zdata, c=zdata)
ax.set_xlabel('Weight')
ax.set_ylabel('Bias')
ax.set_zlabel('loss')
p = ax.scatter(xdata, ydata, zdata, c=zdata)
fig.colorbar(p,pad=0.07)
plt.title('cross entropy loss')
plt.savefig('cross entropy loss')
```

```python
def backprop_entropy(w,b,true_y):
    tmp_1 = np.exp(-(w*x + b))
    tmp_2 = 1 / (1 + np.exp(tmp_1))
    entropy_loss = (tmp_2 - true_y)*x
    return entropy_loss
entropy_loss = backprop_entropy(w,b,true_y)
true_y = np.full((1600, 1), 0.5)
fig = plt.figure()
ax = plt.axes(projection='3d')
zdata = entropy_loss[:,0]
xdata = w[:,0]
ydata = b[:,0]
ax.scatter3D(xdata, ydata, zdata, c=zdata)
ax.set_xlabel('Weight')
ax.set_ylabel('Bias')
ax.set_zlabel('loss')
p = ax.scatter(xdata, ydata, zdata, c=zdata)
fig.colorbar(p,pad=0.07)
plt.title('gradient of cross entropy loss')
plt.savefig('gradient of cross entropy loss')
```
